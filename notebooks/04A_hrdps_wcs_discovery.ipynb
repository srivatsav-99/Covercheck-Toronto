### GetCapabilities and filter HRDPS

```python
from owslib.wcs import WebCoverageService

#GeoMet endpoint
WCS_URL = "https://geo.weather.gc.ca/geomet/"

wcs = WebCoverageService(WCS_URL, version="2.0.1")

all_cov = list(wcs.contents.keys())
print("Total coverages:", len(all_cov))

#filter for HRDPS like coverages
hrdps_cov = [c for c in all_cov if "HRDPS" in c.upper()]
print("HRDPS coverages:", len(hrdps_cov))

for c in hrdps_cov[:100]:
    print(c)
```

### Saving the coverage list

```python
from pathlib import Path
import pandas as pd

REPO_ROOT = Path.cwd().parents[0] if Path.cwd().name == "notebooks" else Path.cwd()
OUT = REPO_ROOT / "docs" / "hrdps_wcs_coverage_list.csv"
OUT.parent.mkdir(parents=True, exist_ok=True)

pd.Series(hrdps_cov, name="coverage_id").to_csv(OUT, index=False)
print("Saved:", OUT)
```
